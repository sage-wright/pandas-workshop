#### Part 2: Data Selection & Indexing
BEHOLD, the Series. 

```python
series = pd.Series([3, 62, 75, 83, 47, 43, 39, 16, 19, 2])
series
```

It's kind of like an array, except it's definitely not an array. It's a Series. 

Guess how the indexing works? Like every single other indexing you've ever used in Python before. 
As a refresher, here's how it goes: 

`series[index]` and `series[print this index : print until this index]` 

OR:

`series.iloc[print this index : print until this index]` Note the [] instead of the ()

This last one has some special powers that only come into play in a little bit, which I will show you soon. 
But what's the difference between `series[4:5]` and `series.iloc[4:5]`, I hear you ask? Well, let me show you.

```python
series[4:5]
```

```python
series.iloc[4:5]
```

_Nothing_. So far, there appears to be no difference. But wait! There's more!

Sometimes you really want to label your indexes for easy access. Sure, I'll buy in. Let's do that ourselves. 

```python
series.index = [x for x in "ABCDEFGHIJKLMNOPQRSTUVWXYZ"][:len(series)]
series
```

Now, instead of numbers for row names, we have LETTERS! Alphabetical letters, in fact. Can these be used to index just the same as numbers? In fact, yes. 

```python
series['D']
```

```python
series['D':'F'] # watch out! when slicing by LABELS, Pandas INCLUDES the end item
```

Can we still index by numbers, even though we have labels? Why, yes we can. 

```python
series[3]
```

```python
series[3:5] # watch out! when slicing by INDEX, Pandas EXCLUDES the end itme
```

What about multiple ranges? Can we slice multiple ranges? Like say, extract D:F and I:J? Let's try!

```python
# by label
series[['D':'F', 'I':'J']]
```

Well, no. We can't by label. What about by index?

```python
# by index
series[[3:5, 8:9]]
```

Seems we can't either. So what can we do??????!? Aha! You wouldn't be bringing this up if you didn't have a solution, you say. How wise you are! This is how we do it:

Use `pd.concat` to combine multiple ranges. 

```python
pd.concat([series['D':'F'], series['I':'J']]) 
# pd.concat([series[3:5], series[8:9]])
# this also works for indexes as well!
```

Here's some words of wisdom regarding labels: 

** if there are duplicate labels, you cannot use them for slicing; i.e., non-unique values breaks slicing, while unique values are still slicable in a non-unique index

Don't believe me? Try out this labeling for fun:

```python
series.index = [x for x in "GATTACAXYZ"][:len(series)]
```

Here are some suggested commands, feel free to use them at your leisure to play around.
   
    series
    series.loc['D']
    series.loc['D':'A'] # non-unique values breaks slicing
    series.loc['X':'Z']  # while unique values are still slicable in a non-unique index
    

**Watch out!** We're using *loc* not *iloc* in this instance. Why do you think that is?
